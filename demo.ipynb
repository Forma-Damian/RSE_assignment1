```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import approximate_taylor_polynomial
from hello_numba.taylor_expansion import math_func

def visualize_taylor(f, x0, max_degree):
    x = np.linspace(-10.0, 10.0, num=100)
    plt.figure(figsize=(10,6))
    plt.plot(x, f(x+x0), label="Original Curve (sin)", color="black", linewidth=2.5)

    # plot each approximation
    for deg in np.arange(1, max_degree + 1, step=2):
        curve_taylor = approximate_taylor_polynomial(f, x0, deg, 1, order=deg + 2)
        plt.plot(x, curve_taylor(x), label=f"Degree={deg}", linestyle="--")

    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.axis([-10, 10, -5, 5])
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.title(f"Taylor Expansion Approximation centered at x={x0}")
    plt.tight_layout()
    plt.show()

# Run the visualization
visualize_taylor(math_func, np.pi, 9)
```
